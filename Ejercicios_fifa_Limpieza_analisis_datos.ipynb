## Cambia las medidas de las variables Weight y Height a Kg y cm

```python
# Observamos como están los valores que queremos cambiar:

df_fifa['Weight'].head
```

```python
# Primero eliminamos las unidades:

df_fifa['Weight_kg'] =  df_fifa['Weight'].str.extract('(\d+)').astype(int)
    
df_fifa['Weight_kg']
```

```python
# Ahora queremos pasar de libras a kg
# kg = lb * 0.45359237

def weight_transform(x):
    return x * 0.45359237

df_fifa['Weight_kg'] = df_fifa['Weight_kg'].apply(weight_transform)

df_fifa['Weight_kg']
```

```python
# Volvemos a observar como están los valores que queremos cambiar, en este caso Height:

df_fifa['Height'].head
```

```python
# Creamos una función para pasar de pies y pulgadas a centimetros:

def height_transform(x):
    if isinstance(x, str):  # Verificamos que sea un string
        x = x.replace('"', '')  # Eliminamos la comilla doble si está presente
        feet, inches = map(int, x.split("'"))  # Separar pies y pulgadas
        return (feet * 12 + inches) * 2.54  # Convertir a cm
    return x  # Si no es un string, lo dejamos igual

df_fifa['Height_cm'] = df_fifa['Height'].apply(height_transform)

df_fifa['Height_cm']
```

## Elimina el "+1", "+2"... de las variables CDM, RDM... y todas las columnas que están al final, sin tener que ir una por una. Apóyate en métodos como split, replace

```python
# Primero busco las columnas que tienen un +algo

def find_columns(df_fifa):
    return [col for col in df_fifa.columns if df_fifa[col].astype(str).str.contains(r'\+').any()]
    
print(find_columns(df_fifa))
```

```python
# Creo una función que separe los numeros de los + o - algo y los elimine
# Aplico la función a las columnas que me interesan

def eliminar_plus(df_fifa):
    for col in df_fifa.columns:
        if df_fifa[col].astype(str).str.contains(r'\+').any():
            df_fifa[col] = df_fifa[col].str.replace(r'[\+\-]\d*', '', regex=True)  # Eliminar números con '+' o '-' algo
            df_fifa[col] = df_fifa[col].str.replace(r'[\+\-]$', '', regex=True)  # Eliminar '+' o '-' al final
            df_fifa[col] = pd.to_numeric(df_fifa[col], errors='coerce') # Convierto a numericas las columnas
    return df_fifa

print(eliminar_plus(df_fifa))
```

## Agrupar o normalizar las posiciones (BP, Best Position) en Portero, Defensa, Mediocampista y delantero. Quizás lo mejor sería crear una función posiciones_jugadores y dependiendo de la BP de cada uno que la nueva variable tome uno de esos 4 valores.

#### GK: Portero, RB: Defensa, CB: Defensa, LB: Defensa, CM: Mediocampista, CAM: Mediocampista, CDM: Mediocampista, LM: Mediocampista, RM: Mediocampista, LWB: Mediocampista, RWB: Mediocampista, ST: Delantero, RW: Delantero, LW: Delantero, CF: Delantero

```python
# Visualizo la columna BP:

df_fifa['BP']
```

```python
# creo la función con las diferentes clasificaciones y almaceno los resultados en una columna nueva:

def posiciones_jugadores(x):
    if x == 'GK':
        return 'Portero'
    elif x in ('RB', 'CB', 'LB'):
        return 'Defensa'
    elif x in ('CM', 'CAM', 'CDM', 'LM', 'RM', 'LWB', 'RWB'):
        return 'Mediocampista'
    elif x in ('ST', 'RW', 'LW', 'CF'):
        return 'Delantero'
    return x 

df_fifa['BP_new'] = df_fifa['BP'].apply(posiciones_jugadores)

df_fifa[['BP', 'BP_new']]
```

## Transforma las variables Value y Wage de manera que sea un entero y podamos calcular la suma y la media por equipo, posición, nacionalidad...

```python
# Visualizo las columnas Value y Wage:

df_fifa[['Value', 'Wage']]
```

```python
# Creo la función y la aplico a las columnas:

def transform_int(x):
    if 'M' in x:
        return float(x.strip('€M'))*1000000 # los valores que son millones multiplicarlos por 1000000 y eliminar el símbolo de €
    elif 'K' in x:
        return float(x.strip('€K'))*1000 # los valores que son miles multiplicarlos por 1000 y eliminar el símbolo de €
    else:
        return float(x.strip('€')) # si no están ninguna de las unidades anteriores solo eliminar el símbolo de €

df_fifa['Value_new'] = df_fifa['Value'].apply(transform_int) # almaceno los resultados en una variable nueva
df_fifa['Wage_new'] = df_fifa['Wage'].apply(transform_int) # almaceno los resultados en una variable nueva

df_fifa[['Value_new', 'Wage_new']]
```
